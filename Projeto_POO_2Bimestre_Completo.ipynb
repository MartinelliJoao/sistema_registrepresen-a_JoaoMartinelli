## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**
- Criar pelo menos 2 m√©todos de a√ß√£o para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
# Classe principal Aluno, com m√©todos de a√ß√£o
class Aluno:
    def __init__(self, nome, matricula, turma):
        self.nome = nome              # Guarda o nome do aluno no atributo 'nome'
        self.matricula = matricula    # Guarda a matr√≠cula do aluno no atributo 'matricula'
        self.turma = turma            # Guarda a turma do aluno no atributo 'turma'
        self.presencas = []           # Inicializa uma lista vazia para armazenar datas de presen√ßa

    # M√©todo para exibir dados do aluno
    def exibir_dados(self):
        print(f"Nome: {self.nome}")         # Exibe o nome do aluno
        print(f"Matr√≠cula: {self.matricula}") # Exibe a matr√≠cula do aluno
        print(f"Turma: {self.turma}")       # Exibe a turma do aluno

    # M√©todo para adicionar presen√ßa (registrar data)
    def registrar_presenca(self, data):
        self.presencas.append(data)                  # Adiciona a data informada √† lista de presen√ßas
        print(f"Presen√ßa registrada para {self.nome} na data {data}")  # Confirma o registro da presen√ßa

    # M√©todo para exibir todas as presen√ßas registradas
    def exibir_presencas(self):
        if self.presencas:                           # Verifica se h√° alguma presen√ßa registrada
            print(f"Presen√ßas de {self.nome}:")     # Exibe o cabe√ßalho das presen√ßas do aluno
            for data in self.presencas:              # Percorre todas as datas registradas
                print(f"- {data}")                   # Exibe cada data
        else:
            print(f"Nenhuma presen√ßa registrada para {self.nome}")  # Mensagem caso n√£o haja presen√ßas registradas

# Subclasse EstudanteGraduacao que herda de Aluno
class EstudanteGraduacao(Aluno):
    def __init__(self, nome, matricula, turma, curso):
        super().__init__(nome, matricula, turma)  # Chama o construtor da classe pai para inicializar os atributos b√°sicos
        self.curso = curso                         # Atributo exclusivo da subclasse que guarda o curso do estudante

    # Novo m√©todo exclusivo da subclasse
    def exibir_curso(self):
        print(f"{self.nome} est√° matriculado no curso de {self.curso}")  # Exibe o curso do estudante

# Testando com inputs para criar um objeto EstudanteGraduacao e registrar presen√ßa

nome = input("Nome do estudante de gradua√ß√£o: ")         # Solicita o nome do estudante
matricula = input("Matr√≠cula: ")                          # Solicita a matr√≠cula
turma = input("Turma: ")                                  # Solicita a turma
curso = input("Curso: ")                                  # Solicita o curso

estudante = EstudanteGraduacao(nome, matricula, turma, curso)  # Cria o objeto EstudanteGraduacao com os dados informados

estudante.exibir_dados()           # Exibe os dados b√°sicos do estudante (nome, matr√≠cula, turma)
estudante.exibir_curso()           # Exibe o curso do estudante

# Registrar uma presen√ßa
data_presenca = input("Digite a data da presen√ßa (ex: 2025-06-02): ")  # Solicita a data da presen√ßa
estudante.registrar_presenca(data_presenca)                            # Registra a presen√ßa do estudante na data informada

# Mostrar as presen√ßas
estudante.exibir_presencas()        # Exibe todas as presen√ßas registradas para o estudante
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
# Classe principal Aluno com encapsulamento e l√≥gica interna
class Aluno:
    def __init__(self, nome, matricula, turma, total_aulas):
        self.nome = nome                    # Guarda o nome do aluno
        self.matricula = matricula          # Guarda a matr√≠cula do aluno
        self.turma = turma                  # Guarda a turma do aluno
        self.__presencas = []               # Atributo privado: lista de datas de presen√ßa
        self.total_aulas = total_aulas      # Total de aulas para c√°lculo da frequ√™ncia

    def get_presencas(self):
        return self.__presencas             # Retorna a lista privada de presen√ßas

    def set_presencas(self, presencas):
        if isinstance(presencas, list):    # Verifica se o valor passado √© uma lista
            self.__presencas = presencas   # Atualiza a lista privada de presen√ßas
        else:
            print("Erro: Presen√ßas devem ser uma lista")  # Mensagem de erro se n√£o for lista

    def registrar_presenca(self, data):
        self.__presencas.append(data)      # Adiciona uma data √† lista privada de presen√ßas
        print(f"Presen√ßa registrada para {self.nome} na data {data}")  # Confirma√ß√£o da presen√ßa

    def calcular_frequencia(self):
        if self.total_aulas == 0:           # Evita divis√£o por zero no c√°lculo
            return 0                        # Retorna zero se total de aulas for zero
        frequencia = (len(self.__presencas) / self.total_aulas) * 100  # Calcula % de presen√ßas
        return frequencia                   # Retorna a frequ√™ncia calculada

    def exibir_dados(self):
        print(f"Nome: {self.nome}")         # Exibe o nome do aluno
        print(f"Matr√≠cula: {self.matricula}")  # Exibe a matr√≠cula do aluno
        print(f"Turma: {self.turma}")       # Exibe a turma do aluno
        print(f"Frequ√™ncia: {self.calcular_frequencia():.2f}%")  # Exibe a frequ√™ncia com 2 casas decimais

# Pedindo dados ao usu√°rio
nome = input("Nome do aluno: ")            # Recebe o nome do aluno
matricula = input("Matr√≠cula: ")            # Recebe a matr√≠cula do aluno
turma = input("Turma: ")                     # Recebe a turma do aluno
total_aulas = int(input("Total de aulas no per√≠odo: "))  # Recebe o total de aulas e converte para inteiro

aluno = Aluno(nome, matricula, turma, total_aulas)  # Cria o objeto Aluno com os dados fornecidos

while True:
    resp = input("Registrar presen√ßa? (s/n): ").lower()  # Pergunta se deseja registrar presen√ßa
    if resp == 's':                                       # Se sim, pede a data
        data = input("Digite a data da presen√ßa (ex: 2025-06-02): ")
        aluno.registrar_presenca(data)                    # Registra a presen√ßa para a data dada
    else:
        break                                            # Sai do la√ßo se a resposta n√£o for 's'

aluno.exibir_dados()                                   # Exibe os dados do aluno e a frequ√™ncia
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
import sqlite3  # Importa a biblioteca para manipular banco de dados SQLite

# Define a classe principal Aluno
class Aluno:
    def __init__(self, nome, matricula, turma):
        self.nome = nome              # Atributo para guardar o nome do aluno
        self.matricula = matricula    # Atributo para guardar a matr√≠cula do aluno
        self.turma = turma            # Atributo para guardar a turma do aluno

    def exibir_dados(self):
        print(f"Nome: {self.nome}")          # Exibe o nome do aluno
        print(f"Matr√≠cula: {self.matricula}")# Exibe a matr√≠cula do aluno
        print(f"Turma: {self.turma}")        # Exibe a turma do aluno

# Define a subclasse EstudanteGraduacao que herda da classe Aluno
class EstudanteGraduacao(Aluno):
    def __init__(self, nome, matricula, turma, curso):
        super().__init__(nome, matricula, turma)  # Chama o construtor da classe pai para inicializar atributos
        self.curso = curso                         # Atributo exclusivo da subclasse para armazenar o curso

    def exibir_dados(self):
        super().exibir_dados()               # Chama o m√©todo exibir_dados da classe pai para mostrar dados b√°sicos
        print(f"Curso: {self.curso}")        # Exibe o curso do estudante (sobrescrevendo o m√©todo, polimorfismo)

# Define a classe BancoDados para manipular banco SQLite
class BancoDados:
    def __init__(self, nome_db="presenca_alunos.db"):
        self.conn = sqlite3.connect(nome_db)   # Estabelece conex√£o com o banco de dados (cria se n√£o existir)
        self.cursor = self.conn.cursor()        # Cria o cursor para executar comandos SQL
        self.criar_tabela()                      # Chama o m√©todo para criar tabela, caso n√£o exista

    def criar_tabela(self):  # Comando SQL para criar a tabela 'alunos' com colunas para id, nome, matr√≠cula, turma e curso
        self.cursor.execute("""
            CREATE TABLE IF NOT EXISTS alunos (
                id INTEGER PRIMARY KEY AUTOINCREMENT,  -- Identificador √∫nico auto-incrementado
                nome TEXT NOT NULL,                      -- Campo para nome do aluno (n√£o pode ser nulo)
                matricula TEXT NOT NULL,                 -- Campo para matr√≠cula do aluno (n√£o pode ser nulo)
                turma TEXT NOT NULL,                     -- Campo para turma (n√£o pode ser nulo)
                curso TEXT                              -- Campo para curso (pode ser nulo)
            )
        """)
        self.conn.commit()  # Aplica as altera√ß√µes no banco de dados

    def inserir_aluno(self, aluno):   # Insere um registro de aluno no banco. O curso pode estar presente ou n√£o (None)
        curso = getattr(aluno, 'curso', None)   # Tenta obter o atributo 'curso', se n√£o existir retorna None
        self.cursor.execute("""
            INSERT INTO alunos (nome, matricula, turma, curso) VALUES (?, ?, ?, ?)
        """, (aluno.nome, aluno.matricula, aluno.turma, curso))  # Executa o comando de inser√ß√£o com valores fornecidos
        self.conn.commit()  # Confirma a opera√ß√£o no banco de dados

    def listar_alunos(self):
        self.cursor.execute("SELECT * FROM alunos")  # Executa comando para buscar todos os alunos
        return self.cursor.fetchall()  # Retorna todos os resultados da consulta como lista de tuplas

    def deletar_aluno(self, matricula):
        self.cursor.execute("DELETE FROM alunos WHERE matricula = ?", (matricula,))  # Deleta aluno pela matr√≠cula
        self.conn.commit()  # Aplica a exclus√£o no banco de dados

    def fechar(self):
        self.conn.close()  # Fecha a conex√£o com o banco de dados

# --- C√≥digo para testar as funcionalidades ---

banco = BancoDados()  # Cria o objeto que gerencia o banco de dados SQLite

nome = input("Nome do estudante de gradua√ß√£o: ")    # Recebe o nome do estudante
matricula = input("Matr√≠cula: ")                     # Recebe a matr√≠cula
turma = input("Turma: ")                             # Recebe a turma
curso = input("Curso: ")                             # Recebe o curso

estudante = EstudanteGraduacao(nome, matricula, turma, curso)  # Cria um objeto EstudanteGraduacao com os dados

banco.inserir_aluno(estudante)  # Insere o aluno no banco de dados

print("\nDados do estudante inserido:")
estudante.exibir_dados()  # Exibe os dados do estudante usando o m√©todo sobrescrito

print("\nLista de todos os alunos no banco:")
alunos = banco.listar_alunos()  # Busca todos os alunos no banco
for a in alunos:
    print(a)  # Exibe cada registro como uma tupla (id, nome, matr√≠cula, turma, curso)

resp = input("\nDeseja excluir algum aluno pelo n√∫mero da matr√≠cula? (s/n): ").lower()  # Pergunta se deseja excluir aluno
if resp == 's':
    matricula_del = input("Digite a matr√≠cula do aluno a ser exclu√≠do: ")  # Pede a matr√≠cula para exclus√£o
    banco.deletar_aluno(matricula_del)  # Executa a exclus√£o
    print("Aluno exclu√≠do.")

banco.fechar()  # Fecha a conex√£o com o banco
```

## PARTE 5 ‚Äì Projeto Final e Organiza√ß√£o

**Tarefas:**
- Consolidar todas as partes do projeto em um √∫nico script organizado.
- Comentar o c√≥digo explicando cada parte.
- Testar todas as funcionalidades.

**C√≥digo:**

```python
import sqlite3  # Biblioteca para manipula√ß√£o do banco SQLite

# === Classe principal para gerenciar alunos ===
class Aluno:
    def __init__(self, nome, matricula, turma):
        self.__nome = nome             # Atributo privado para nome do aluno
        self.__matricula = matricula   # Atributo privado para matr√≠cula
        self.__turma = turma           # Atributo privado para turma
        self.__presencas = []          # Lista privada para armazenar datas de presen√ßa

    # M√©todo getter para nome
    def get_nome(self):
        return self.__nome

    # M√©todo setter para nome
    def set_nome(self, novo_nome):
        self.__nome = novo_nome

    # M√©todo getter para matr√≠cula
    def get_matricula(self):
        return self.__matricula

    # M√©todo setter para matr√≠cula
    def set_matricula(self, nova_matricula):
        self.__matricula = nova_matricula

    # M√©todo getter para turma
    def get_turma(self):
        return self.__turma

    # M√©todo setter para turma
    def set_turma(self, nova_turma):
        self.__turma = nova_turma

    # M√©todo para exibir dados do aluno
    def exibir_dados(self):
        print(f"Nome: {self.__nome}")
        print(f"Matr√≠cula: {self.__matricula}")
        print(f"Turma: {self.__turma}")

    # M√©todo para registrar presen√ßa em uma data
    def registrar_presenca(self, data):
        self.__presencas.append(data)
        print(f"Presen√ßa registrada para {self.__nome} na data {data}")

    # M√©todo para exibir todas as presen√ßas registradas
    def exibir_presencas(self):
        if self.__presencas:
            print(f"Presen√ßas de {self.__nome}:")
            for data in self.__presencas:
                print(f"- {data}")
        else:
            print(f"Nenhuma presen√ßa registrada para {self.__nome}")

# === Subclasse para estudantes de gradua√ß√£o ===
class EstudanteGraduacao(Aluno):
    def __init__(self, nome, matricula, turma, curso):
        super().__init__(nome, matricula, turma)  # Inicializa atributos da classe pai
        self.__curso = curso                       # Atributo privado para curso

    # Getter para curso
    def get_curso(self):
        return self.__curso

    # Setter para curso
    def set_curso(self, novo_curso):
        self.__curso = novo_curso

    # Sobrescreve m√©todo para exibir dados, adicionando curso
    def exibir_dados(self):
        super().exibir_dados()                # Exibe dados da classe pai
        print(f"Curso: {self.__curso}")       # Exibe o curso do estudante

# === Classe para manipula√ß√£o do banco de dados SQLite ===
class BancoDados:
    def __init__(self, nome_db="presenca_alunos.db"):
        self.conn = sqlite3.connect(nome_db)   # Conecta/cria banco SQLite
        self.cursor = self.conn.cursor()        # Cursor para executar comandos SQL
        self.criar_tabela()                      # Garante que a tabela exista

    # Cria tabela 'alunos' caso n√£o exista
    def criar_tabela(self):
        self.cursor.execute("""
            CREATE TABLE IF NOT EXISTS alunos (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nome TEXT NOT NULL,
                matricula TEXT NOT NULL UNIQUE,
                turma TEXT NOT NULL,
                curso TEXT
            )
        """)
        self.conn.commit()

    # Insere um aluno no banco (pode ser Aluno ou EstudanteGraduacao)
    def inserir_aluno(self, aluno):
        curso = None
        if isinstance(aluno, EstudanteGraduacao):
            curso = aluno.get_curso()
        self.cursor.execute("""
            INSERT INTO alunos (nome, matricula, turma, curso) VALUES (?, ?, ?, ?)
        """, (aluno.get_nome(), aluno.get_matricula(), aluno.get_turma(), curso))
        self.conn.commit()
        print(f"Aluno {aluno.get_nome()} inserido no banco.")

    # Lista todos os alunos
    def listar_alunos(self):
        self.cursor.execute("SELECT * FROM alunos")
        return self.cursor.fetchall()

    # Deleta um aluno pelo n√∫mero da matr√≠cula
    def deletar_aluno(self, matricula):
        self.cursor.execute("DELETE FROM alunos WHERE matricula = ?", (matricula,))
        self.conn.commit()
        print(f"Aluno com matr√≠cula {matricula} deletado.")

    # Fecha a conex√£o com o banco
    def fechar(self):
        self.conn.close()

# === Programa principal para intera√ß√£o com o usu√°rio ===
def main():
    banco = BancoDados()  # Cria conex√£o com banco SQLite

    while True:
        print("\n=== Menu ===")
        print("1 - Inserir estudante de gradua√ß√£o")
        print("2 - Listar todos os alunos")
        print("3 - Deletar aluno por matr√≠cula")
        print("4 - Sair")
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == '1':
            nome = input("Nome do estudante: ")
            matricula = input("Matr√≠cula: ")
            turma = input("Turma: ")
            curso = input("Curso: ")
            estudante = EstudanteGraduacao(nome, matricula, turma, curso)
            banco.inserir_aluno(estudante)

        elif opcao == '2':
            alunos = banco.listar_alunos()
            print("\nAlunos cadastrados:")
            for a in alunos:
                print(f"ID: {a[0]}, Nome: {a[1]}, Matr√≠cula: {a[2]}, Turma: {a[3]}, Curso: {a[4]}")

        elif opcao == '3':
            matricula = input("Digite a matr√≠cula do aluno para deletar: ")
            banco.deletar_aluno(matricula)

        elif opcao == '4':
            banco.fechar()
            print("Programa encerrado.")
            break

        else:
            print("Op√ß√£o inv√°lida, tente novamente.")

# Executa o programa principal se o script for rodado diretamente
if __name__ == "__main__":
    main()
```
