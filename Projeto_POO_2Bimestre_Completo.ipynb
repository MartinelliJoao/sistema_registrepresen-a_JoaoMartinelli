## PARTE 1 ‚Äì Classe Principal e Objetos

**Tarefas:**
- Definir e criar a classe principal com pelo menos 3 atributos.
- Implementar o m√©todo construtor (`__init__`).
- Criar pelo menos 2 objetos da classe principal.
- Implementar um m√©todo de exibi√ß√£o de dados.

**C√≥digo:**

```python
class Presenca:  # Define a classe Presenca para armazenar dados de presen√ßa
    def __init__(self, nome, data, presente):  # M√©todo construtor que inicializa o objeto
        self.nome = nome          # Atribui o nome do aluno ao objeto
        self.data = data          # Atribui a data ao objeto
        self.presente = presente  # Atribui se o aluno est√° presente (True ou False)
        
    def exibir_dados(self):  # M√©todo para mostrar os dados do objeto
        status = "Presente" if self.presente else "Ausente"  # Define o status como texto
        print(f"Nome: {self.nome}")     # Imprime o nome do aluno
        print(f"Data: {self.data}")     # Imprime a data
        print(f"Status: {status}")      # Imprime o status (Presente ou Ausente)


def criar_presenca():  # Fun√ß√£o que coleta dados do usu√°rio e cria um objeto Presenca
    nome = input("Digite o nome do aluno: ")  # Solicita o nome do aluno
    data = input("Digite a data (ex: 2025-05-11): ")  # Solicita a data
    
    while True:  # Loop para garantir que o usu√°rio informe 's' ou 'n' corretamente
        presente_input = input("O aluno est√° presente? (s/n): ").strip().lower()  # Pede presen√ßa
        if presente_input in ['s', 'n']:  # Se resposta for v√°lida (sim ou n√£o)
            presente = True if presente_input == 's' else False  # Converte para booleano
            break  # Sai do loop
        else:
            print("Por favor, responda com 's' para sim ou 'n' para n√£o.")  # Mensagem de erro para resposta inv√°lida
    
    return Presenca(nome, data, presente)  # Retorna o objeto Presenca criado com os dados


# Exemplo de uso criando e exibindo dados de dois alunos

aluno1 = criar_presenca()  # Cria o primeiro aluno pedindo dados do usu√°rio
print()  # Linha em branco para separar no terminal
aluno2 = criar_presenca()  # Cria o segundo aluno
print()

aluno1.exibir_dados()  # Mostra os dados do primeiro aluno
print()
aluno2.exibir_dados()  # Mostra os dados do segundo aluno
```

## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**
- Criar pelo menos 2 m√©todos de a√ß√£o para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
# Classe principal para controlar presen√ßa de alunos
class Presenca:
    # Construtor que inicializa nome, data e status de presen√ßa
    def __init__(self, nome, data, presente):
        self.nome = nome              # Atributo nome do aluno
        self.data = data              # Atributo data da aula
        self.presente = presente      # Atributo booleano: True se presente, False se ausente

    # M√©todo para exibir os dados do aluno e seu status
    def exibir_dados(self):
        status = "Presente" if self.presente else "Ausente"  # Define texto do status
        print(f"Nome: {self.nome}")
        print(f"Data: {self.data}")
        print(f"Status: {status}")

    # M√©todo para marcar presen√ßa (a√ß√£o que muda o estado do objeto)
    def marcar_presenca(self):
        self.presente = True  # Marca o aluno como presente

    # M√©todo para justificar falta (a√ß√£o condicionada pelo estado)
    def justificar_falta(self):
        if not self.presente:  # S√≥ justifica se o aluno est√° ausente
            print(f"{self.nome} justificou a falta no dia {self.data}.")
        else:
            print(f"{self.nome} estava presente, n√£o precisa justificar.")


# Subclasse que herda da classe Presenca e adiciona novo atributo e m√©todo
class PresencaDetalhada(Presenca):
    # Construtor da subclasse que usa super() para reaproveitar o construtor da classe pai
    def __init__(self, nome, data, presente, motivo):
        super().__init__(nome, data, presente)  # Chama o construtor da classe Presenca
        self.motivo = motivo                    # Novo atributo exclusivo da subclasse

    # Novo m√©todo exclusivo que exibe detalhes, incluindo motivo da falta se ausente
    def exibir_detalhes(self):
        self.exibir_dados()  # Reusa m√©todo da classe pai para exibir dados b√°sicos
        if not self.presente:
            print(f"Motivo da falta: {self.motivo}")  # Exibe motivo da falta


# Criando inst√¢ncias para testar as classes e seus m√©todos
aluno1 = Presenca("Jo√£o Pedro", "2025-05-11", True)  # Objeto da classe principal
aluno2 = PresencaDetalhada("Maria Clara", "2025-05-11", False, "Consulta m√©dica")  # Objeto da subclasse

aluno1.exibir_dados()        # Exibe dados do aluno1
print()
aluno2.exibir_detalhes()     # Exibe dados detalhados do aluno2
print()
aluno2.justificar_falta()    # Testa m√©todo herdado para justificar falta
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
class Presenca:
    def __init__(self, nome, data, presente):
        self.__nome = nome        # Atributo privado para proteger o nome
        self.data = data          # Atributo p√∫blico data
        self.presente = presente  # Atributo p√∫blico presen√ßa (True/False)

    def get_nome(self):           # M√©todo getter para acessar o nome privado
        return self.__nome

    def set_nome(self, novo_nome):  # M√©todo setter para alterar o nome com valida√ß√£o
        if isinstance(novo_nome, str) and novo_nome.strip():  # Verifica se √© string n√£o vazia
            self.__nome = novo_nome
        else:
            print("Nome inv√°lido.")  # Mensagem de erro para nome inv√°lido

    def exibir_dados(self):         # Exibe informa√ß√µes do aluno e presen√ßa
        status = "Presente" if self.presente else "Ausente"
        print(f"Nome: {self.__nome}")
        print(f"Data: {self.data}")
        print(f"Status: {status}")

    def verificar_falta(self):      # Retorna True se aluno est√° ausente
        return not self.presente

class PresencaDetalhada(Presenca):
    def __init__(self, nome, data, presente, motivo):
        super().__init__(nome, data, presente)  # Reaproveita construtor da classe pai
        self.motivo = motivo                    # Novo atributo exclusivo da subclasse

    def exibir_detalhes(self):     # Mostra dados e motivo da falta se houver
        self.exibir_dados()
        if self.verificar_falta():
            print(f"Motivo da falta: {self.motivo}")

aluno1 = Presenca("Jo√£o Pedro", "2025-05-11", True)               # Inst√¢ncia da classe base
aluno2 = PresencaDetalhada("Maria Clara", "2025-05-11", False, "Consulta m√©dica")  # Inst√¢ncia da subclasse

aluno1.exibir_dados()       # Exibe dados do aluno1
print()
aluno2.exibir_detalhes()    # Exibe dados e motivo do aluno2
print()
aluno1.set_nome("Jo√£o P.")  # Atualiza o nome de aluno1 via setter
print("Novo nome:", aluno1.get_nome())  # Imprime o nome atualizado via getter
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
import sqlite3  # Importa o m√≥dulo para trabalhar com banco de dados SQLite

# Classe para controlar presen√ßa b√°sica
class Presenca:
    def __init__(self, nome, data, presente):
        self.__nome = nome          # Atributo privado para o nome
        self.data = data            # Data da presen√ßa
        self.presente = presente    # Booleano que indica presen√ßa (True/False)

    def get_nome(self):            # M√©todo para acessar o nome privado
        return self.__nome

    def set_nome(self, novo_nome):    # M√©todo para alterar o nome com valida√ß√£o
        if isinstance(novo_nome, str) and novo_nome.strip():
            self.__nome = novo_nome
        else:
            print("Nome inv√°lido.")

    def exibir_dados(self):       # Exibe os dados b√°sicos da presen√ßa
        status = "Presente" if self.presente else "Ausente"
        print(f"[B√ÅSICO] Nome: {self.__nome}")
        print(f"Data: {self.data}")
        print(f"Status: {status}")

    def verificar_falta(self):    # Retorna True se o aluno est√° ausente
        return not self.presente

# Subclasse que adiciona motivo da falta
class PresencaDetalhada(Presenca):
    def __init__(self, nome, data, presente, motivo):
        super().__init__(nome, data, presente)  # Chama o construtor da classe base
        self.motivo = motivo                    # Novo atributo motivo da falta

    def exibir_dados(self):  # Sobrescreve o m√©todo para mostrar motivo se ausente
        super().exibir_dados()
        if self.verificar_falta():
            print(f"[DETALHADO] Motivo da falta: {self.motivo}")

# Fun√ß√£o para criar a tabela no banco, caso n√£o exista
def criar_tabela():
    conexao = sqlite3.connect("presencas.db")  # Abre conex√£o com o banco
    cursor = conexao.cursor()                    # Cria cursor para executar comandos
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS presenca (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            nome TEXT,
            data TEXT,
            presente INTEGER
        )
    """)  # Cria a tabela 'presenca' com colunas id, nome, data e presente
    conexao.commit()    # Confirma a cria√ß√£o da tabela
    conexao.close()     # Fecha a conex√£o com o banco

# Fun√ß√£o para inserir um registro de presen√ßa na tabela
def inserir_presenca(nome, data, presente):
    conexao = sqlite3.connect("presencas.db")  # Abre conex√£o
    cursor = conexao.cursor()
    cursor.execute("INSERT INTO presenca (nome, data, presente) VALUES (?, ?, ?)", 
                   (nome, data, 1 if presente else 0))  # Insere dados, convertendo True/False para 1/0
    conexao.commit()  # Confirma inser√ß√£o
    conexao.close()   # Fecha conex√£o

# Fun√ß√£o para listar todas as presen√ßas armazenadas
def listar_presencas():
    conexao = sqlite3.connect("presencas.db")
    cursor = conexao.cursor()
    cursor.execute("SELECT * FROM presenca")  # Seleciona todos os registros
    resultados = cursor.fetchall()            # Busca todos os resultados
    for id, nome, data, presente in resultados:
        status = "Presente" if presente == 1 else "Ausente"  # Converte 1/0 em texto
        print(f"ID: {id} | Nome: {nome} | Data: {data} | Status: {status}")
    conexao.close()

# Fun√ß√£o para deletar um registro pelo seu ID
def deletar_presenca(id):
    conexao = sqlite3.connect("presencas.db")
    cursor = conexao.cursor()
    cursor.execute("DELETE FROM presenca WHERE id = ?", (id,))  # Deleta registro com o id informado
    conexao.commit()
    conexao.close()

# Execu√ß√£o do programa
criar_tabela()  # Garante que a tabela exista

# Insere dois registros de presen√ßa
inserir_presenca("Jo√£o Pedro", "2025-05-11", True)
inserir_presenca("Maria Clara", "2025-05-11", False)

print("=== Lista de Presen√ßas ===")
listar_presencas()  # Mostra os registros atuais

print("\n=== Deletando presen√ßa com ID 1 ===")
deletar_presenca(1)  # Deleta o registro com id 1

print("\n=== Lista Atualizada ===")
listar_presencas()  # Mostra os registros ap√≥s a exclus√£o
```

**C√≥digo:**

```python
import sqlite3  # Importa o m√≥dulo para trabalhar com banco SQLite

class Presenca:
    def __init__(self, nome, data, presente):
        self.__nome = nome         # Atributo privado para nome
        self.data = data           # Atributo para a data
        self.presente = presente   # Atributo que indica se est√° presente (True/False)

    def get_nome(self):
        return self.__nome         # M√©todo para acessar o nome (getter)

    def set_nome(self, novo_nome):
        # M√©todo para alterar o nome (setter) s√≥ aceita texto v√°lido
        if isinstance(novo_nome, str) and novo_nome.strip():
            self.__nome = novo_nome
        else:
            print("Nome inv√°lido.")  # Se tentar mudar para algo inv√°lido

    def exibir_dados(self):
        # Mostra os dados do aluno: nome, data e status
        status = "Presente" if self.presente else "Ausente"
        print(f"[B√ÅSICO] Nome: {self.__nome}")
        print(f"Data: {self.data}")
        print(f"Status: {status}")

    def verificar_falta(self):
        # Retorna True se o aluno estiver ausente, False se presente
        return not self.presente

class PresencaDetalhada(Presenca):
    def __init__(self, nome, data, presente, motivo):
        super().__init__(nome, data, presente)  # Usa o construtor da classe base
        self.motivo = motivo                     # Novo atributo exclusivo para motivo da falta

    def exibir_dados(self):
        super().exibir_dados()                   # Chama o m√©todo da classe base para mostrar os dados b√°sicos
        if self.verificar_falta():
            print(f"[DETALHADO] Motivo da falta: {self.motivo}")  # Exibe o motivo se estiver ausente

def criar_tabela():
    conexao = sqlite3.connect("presencas.db")  # Abre (ou cria) o banco de dados
    cursor = conexao.cursor()
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS presenca (
            id INTEGER PRIMARY KEY AUTOINCREMENT,  # ID autoincrement√°vel
            nome TEXT,
            data TEXT,
            presente INTEGER
        )
    """)  # Cria a tabela se ela n√£o existir
    conexao.commit()  # Salva as altera√ß√µes
    conexao.close()   # Fecha a conex√£o

def inserir_presenca(nome, data, presente):
    conexao = sqlite3.connect("presencas.db")  
    cursor = conexao.cursor()
    cursor.execute("INSERT INTO presenca (nome, data, presente) VALUES (?, ?, ?)",
                   (nome, data, 1 if presente else 0))  # Insere os dados na tabela, convertendo True/False para 1/0
    conexao.commit()
    conexao.close()

def listar_presencas():
    conexao = sqlite3.connect("presencas.db")
    cursor = conexao.cursor()
    cursor.execute("SELECT * FROM presenca")  # Seleciona todos os registros
    resultados = cursor.fetchall()            # Pega todos os resultados
    for id, nome, data, presente in resultados:
        status = "Presente" if presente == 1 else "Ausente"
        print(f"ID: {id} | Nome: {nome} | Data: {data} | Status: {status}")
    conexao.close()

def deletar_presenca(id):
    conexao = sqlite3.connect("presencas.db")
    cursor = conexao.cursor()
    cursor.execute("DELETE FROM presenca WHERE id = ?", (id,))  # Deleta o registro com o id informado
    conexao.commit()
    conexao.close()

# --- Uso do sistema ---

criar_tabela()  # Garante que a tabela exista

# Cria os objetos aluno1 e aluno2 (um da classe base e outro da subclasse)
aluno1 = Presenca("Jo√£o Pedro", "2025-05-11", True)
aluno2 = PresencaDetalhada("Maria Clara", "2025-05-11", False, "Consulta m√©dica")

# Insere os dados dos alunos no banco, usando os m√©todos get_nome(), data e presente
inserir_presenca(aluno1.get_nome(), aluno1.data, aluno1.presente)
inserir_presenca(aluno2.get_nome(), aluno2.data, aluno2.presente)

print("\n=== Exibi√ß√£o com Polimorfismo ===")
lista_presencas = [aluno1, aluno2]   # Lista de objetos que usa polimorfismo para exibir dados
for aluno in lista_presencas:
    aluno.exibir_dados()              # Cada objeto usa seu pr√≥prio m√©todo exibir_dados
    print()

print("=== Registros no Banco de Dados ===")
listar_presencas()  # Mostra o que est√° salvo no banco

print("\n=== Deletando Registro com ID 1 ===")
deletar_presenca(1)  # Apaga o registro com ID 1

print("\n=== Lista Atualizada ===")
listar_presencas()  # Mostra a lista ap√≥s a exclus√£o
```
